# 🌍 1. İnsan Yerleşimi ile Meteor Düşüşlerinin Coğrafi Yakınlığı

## 🎯 Hipotez

> İnsanlar tarih boyunca meteor düşüşü yaşanan bölgelerde yaşam kurmuştur.  
> Bunun nedeni, meteorların düştüğü bölgelerin genellikle mineral ve maden açısından zengin olması ve bu bölgelerin ekonomik cazibe yaratmasıdır.

Bu hipotez, coğrafi yakınlık analizi ile test edilebilir:
- Eğer şehirlerin büyük bir kısmı meteor düşüşlerine yakınsa → bu ilişki istatistiksel olarak gözlemlenebilir olur.
- Aksi takdirde, şehirler meteor noktalarından rastgele uzaklıklarda olurdu.

---

## 📊 Veri Kaynakları

- `meteorite_landings.csv`: Dünya çapında düşmüş 38.401 adet meteorun konum verisi
- `ne_10m_populated_places.shp`: Dünya genelinde 7.300’den fazla yerleşim biriminin konumu

Bu veriler kullanılarak her şehir için en yakın meteor düşüşü ile olan mesafe hesaplandı.

---

## 📌 Ana Bulgular

- Ortalama şehir, meteora **yaklaşık 365 km** uzaklıkta yer alıyor.
- Tüm şehirlerin **%85.49’u**, en yakın meteora **500 km'den daha kısa bir mesafede** yer alıyor.

Bu sonuçlar, şehir konumlarının meteor düşüş alanlarına rastlantısal değil, **belirli bir yakınlık düzeniyle** konumlandığını göstermektedir.

---

## 📈 Görselleştirmeler

Aşağıdaki grafikler hipotezimizi destekleyen ilk bulguları sunmaktadır:

1. **Histogram**: Tüm şehirlerin meteorlara olan minimum mesafe dağılımı  
2. **Pie Chart**: 500 km'den kısa mesafede olan şehir oranı (%85.49)

Bu grafikler, şehirlerin büyük çoğunluğunun meteor bölgelerine oldukça yakın olduğunu açıkça göstermektedir.


## 🤖 2. Makine Öğrenmesi ile Coğrafi Yakınlık Tahmini

Hipotezimizi daha ileri düzeyde test edebilmek için, şehirlerin meteorlara olan yakınlığını tahmin etmeyi amaçlayan bir dizi makine öğrenmesi modeli uygulanmıştır. Bu süreçte kullanılan bağımsız değişkenler şunlardır:

- `min_meteor_distance_m`: Her şehrin en yakın meteora olan mesafesi (metre cinsinden)
- Kıta bazlı kategorik değişkenler (Europe, Asia, North America vb.)

### 🎯 Amaç

> "Bir şehrin meteora 500 km'den daha yakın olup olmadığını (yakınlık etiketini), sadece konum ve kıta bilgilerine bakarak tahmin edebilir miyiz?"

---

### 📊 Kullanılan Modeller

Aşağıdaki modeller her şehir için `near_meteor` (1 = yakın, 0 = uzak) etiketini tahmin etmek için eğitilmiştir:

| Model               | Accuracy | Precision | Recall | F1 Score |
|---------------------|----------|-----------|--------|----------|
| Logistic Regression | 0.863    | 0.863     | 1.000  | 0.927    |
| Decision Tree       | 0.863    | 0.863     | 1.000  | 0.927    |
| Random Forest       | 0.863    | 0.863     | 1.000  | 0.927    |
| XGBoost             | 0.863    | 0.863     | 1.000  | 0.927    |
| KNN                 | 0.856    | 0.891     | 0.949  | 0.919    |
| Naive Bayes         | 0.856    | 0.891     | 0.949  | 0.919    |

---

### 🧠 Yorum

- Özellikle `Recall` değerinin yüksek çıkması, modellerin **yakın şehirleri tespit etmede oldukça başarılı** olduğunu göstermektedir.
- Tüm modellerin F1 Score değeri %92 seviyelerine yakın olup, bu durum **hem doğru pozitif hem de doğru negatif sınıflandırmalarda denge** olduğunu işaret eder.
- Farklı algoritmaların benzer başarı oranları göstermesi, **coğrafi konumun bu problemde belirleyici bir faktör** olduğunu desteklemektedir.

Bu sonuçlar, hipotezimizi destekleyen önemli bir yapay zeka analizidir.


## 🌐 3. Coğrafi Kümeleme: Meteorlar Belirli Bölgelerde mi Toplanıyor?

Meteor düşüşlerinin dünya üzerinde rastgele mi dağıldığı, yoksa belirli bölgelerde mi yoğunlaştığı hipotezin temel sorularından biridir. Eğer meteorlar **spesifik coğrafi bölgelerde yoğunlaşıyorsa** ve bu bölgeler aynı zamanda insan yerleşimleriyle örtüşüyorsa, bu durum **hipotezi doğrudan güçlendirir.**

---

### 🧪 Yöntem: Kümeleme Algoritmaları

İki farklı algoritma kullanılmıştır:

- **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**  
- **HDBSCAN (Hierarchical DBSCAN)**

Bu algoritmalar, meteorit koordinatlarını analiz ederek doğal yoğunlaşmaları (kümeleri) ortaya çıkarmaktadır.

---

### 📌 Bulgular

- DBSCAN ile yaklaşık **40 anlamlı küme** tespit edilmiştir.
- HDBSCAN ile **310 adet coğrafi küme** saptanmıştır.  
  Bu kümelerin bazıları çok yoğun ve büyük şehir kümeleriyle **coğrafi olarak çakışmaktadır**.
  
Özellikle Avrupa, Hindistan, Kuzey Amerika gibi bölgelerde bu örtüşme daha net gözlenmektedir.

---

### 🗺️ Görsel Harita

Aşağıdaki görsellerde:

- En yoğun 10 HDBSCAN kümesi, modern şehir yerleşimlerine göre gösterilmiştir.
- Küme renkleri açık renklerden seçilmiş, arka plan koyu tutulmuştur.
- Kümelerin merkezleri, şehirlerin bulunduğu bölgelerle örtüşmektedir.

Bu durum, **meteor düşüş bölgeleri ile insan yerleşimleri arasında mekânsal bir bağ** olduğunu göstermektedir.

---

### 🧩 Yorum

> Coğrafi kümelerin oluşması, meteor düşüşlerinin belirli bölgelere odaklandığını ve bu bölgelerin insan yerleşimleriyle çakıştığını gösteriyor.

Bu da hipotezi şu şekilde destekler:  
**İnsanlar tarih boyunca, meteorların yoğunlaştığı maden zengini bölgelerde yerleşmeyi tercih etmiştir.**



## 🕰️ 4. Meteor Gözlemlerinin Zamanla Artışı: Modern Yerleşimlerle Paralellik Gösteriyor mu?

Meteor düşüşlerinin zaman içerisindeki dağılımı da hipotezimizi dolaylı olarak destekleyebilir. Özellikle modern çağda artan şehirleşme ile birlikte meteor düşüş gözlemlerinin de artması, bu iki olgunun **gözlem, kayıt ve yerleşim dinamikleriyle ilişkili** olduğunu gösterebilir.

---

### 📅 Yöntem: Zaman Serisi Analizi

Meteorite verisinde yer alan düşüş yılları kullanılarak, yıllara göre kaç meteorun kayıt altına alındığı analiz edilmiştir.

---

### 📌 Bulgular

- 1800'lü yıllarda yılda sadece birkaç meteor düşüşü kaydedilirken,
- **1950 sonrası** gözlemlerde **ani bir artış** yaşanmıştır.
- Son 50 yılda meteor gözlemlerinde çok daha yoğun bir kayıt süreci gerçekleşmiştir.

---

### 📊 Görsel: Zaman Serisi Grafiği

- Çizgi grafik ile yıllara göre düşen meteor sayısı gösterilmiştir.
- 1950'den itibaren belirgin bir yükseliş eğilimi gözlemlenmiştir.
- Bu artış teknolojik ilerleme (daha iyi gözlem sistemleri), artan şehirleşme ve küresel kayıt sistemlerinin gelişimiyle paraleldir.

---

### 📌 Yorum

> Bu bulgu doğrudan hipotezimizi kanıtlamasa da, **meteor gözlemlerinin şehirleşme ve insan varlığıyla birlikte arttığını** dolaylı olarak göstermektedir.

Bu durum, **meteor düşüşlerinin insan yerleşimleriyle zaman içinde daha fazla kesişme ihtimalini** güçlendirir.


## 🧠 5. Genel Sonuçlar ve Hipotezin Değerlendirilmesi

Bu çalışmada, "İnsanlar tarih boyunca meteor düşüş bölgelerine yakın yerlerde mi yaşam kurmuştur?" hipotezini veri analizi ve makine öğrenmesi ile test ettik.

---

### 🔍 Bulguların Özeti

| Analiz Türü        | Sonuç                                                                 |
|--------------------|------------------------------------------------------------------------|
| 📏 Mesafe Analizi   | Şehirlerin %85.49'u, en yakın meteora 500 km'den daha kısa mesafede   |
| 🤖 ML Modelleri     | %86+ doğrulukla şehirlerin yakın/uzak sınıfı tahmin edildi             |
| 🧭 Kümeleme Analizi | Meteorlar belirli coğrafi bölgelerde kümelenmiş → bu bölgelerle şehirler çakışıyor |
| 📈 Zaman Serisi     | 1950 sonrası meteor gözlemlerinde artış → şehirleşme ve gözlem ilişkisi |

---

### 🎯 Hipotezin Gücü

- **Mekânsal analiz** ve **sınıflandırma modelleri**, şehirlerin meteor bölgelerine rastlantısal olmayan bir şekilde yerleştiğini gösterdi.
- **Yoğun meteor kümeleri** ile **şehirlerin örtüşmesi**, mineral kaynaklarının insan yerleşiminde etkili olabileceğini desteklemektedir.
- **Zamanla artan gözlemler**, şehirleşmenin gözlemlenebilir meteor kayıtlarına olan etkisini ortaya koymuştur.

---

### 📌 Kısıtlamalar

- Meteor verileri sadece “kayıt altına alınabilen” düşüşleri içeriyor olabilir.
- Yerleşim verileri sadece modern dönem şehirlerini kapsamaktadır.
- Coğrafi örtüşme, nedensellik anlamına gelmez; ancak güçlü bir korelasyon gözlenmiştir.

---

### ✅ Sonuç

> Yapılan analizler, insan yerleşimlerinin meteor düşüş noktalarına **coğrafi olarak anlamlı derecede yakın** olduğunu ve bu ilişkinin **istatistiksel modellerle de doğrulanabilir** olduğunu göstermektedir.

Hipotezimiz, hem veri destekli hem de teorik olarak **anlamlı ve güçlü şekilde desteklenmiştir**.
